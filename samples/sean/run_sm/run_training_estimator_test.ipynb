```python
from sagemaker.estimator import Estimator
from sagemaker.session import Session
from sagemaker import get_execution_role
```

```python
role = get_execution_role()
sess = Session()
```

```python
estimator = Estimator(
    image_uri='155954279556.dkr.ecr.us-west-2.amazonaws.com/gs-automl-base-containers/boilerplate312_sm:sean-v3.0',
    role=role,
    instance_count=1,
    instance_type='ml.m5.large',
    hyperparameters={
        "conf-s3-path": "s3://gs-retail-awesome-conf-us-west-2/dev/sample/titanic-survival-prediction/baseline-awesome-sean-v1/",
        "notebook-path": "s3://gs-retail-awesome-conf-us-west-2/dev/sample/titanic-survival-prediction/baseline-awesome-sean-v1/modeling.ipynb",
    },
    base_job_name='sean-sample-training',
    sagemaker_session=sess
)

estimator.fit()
```

```python
# 이후, 최신 트레이닝 잡 정보 접근
training_job_name = estimator.latest_training_job.name
print("Training Job Name:", training_job_name)

# CloudWatch 로그 그룹과 스트림 이름 가져오기
log_group = f"/aws/sagemaker/TrainingJobs"
log_stream = f"{training_job_name}/algo-1"

print("Log Group:", log_group)
print("Log Stream:", log_stream)
```

```python
import boto3

logs_client = boto3.client("logs", region_name="us-west-2")
```

```python
streams = logs_client.describe_log_streams(
    logGroupName="/aws/sagemaker/TrainingJobs",
    logStreamNamePrefix=f"{training_job_name}/algo-1"
)

for stream in streams['logStreams']:
    print(stream['logStreamName'])  # 전체 UUID 포함된 스트림 이름
```
