```python
import boto3

region_name = 'ap-northeast-2'

def fetch_sagemaker_logs(job_name, log_group='/aws/sagemaker/TrainingJobs'):
    logs_client = boto3.client('logs', region_name=region_name)
    log_stream_name = job_name

    #  : orderBy 
    streams = logs_client.describe_log_streams(
        logGroupName=log_group,
        logStreamNamePrefix=log_stream_name
    )

    if not streams['logStreams']:
        print(f"     : {log_stream_name}")
        return []

    #     
    stream = streams['logStreams'][0]['logStreamName']

    #   
    events = logs_client.get_log_events(
        logGroupName=log_group,
        logStreamName=stream,
        startFromHead=True
    )

    logs = [event['message'] for event in events['events']]
    return logs
```

```python
logs = fetch_sagemaker_logs(job_name)
print("\n".join(logs))
```
