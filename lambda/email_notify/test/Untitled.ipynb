```python
import boto3
from botocore.exceptions import ClientError

region_name = 'ap-northeast-2'  # 필요 시 수정

def put_parameter_key(param_name: str, param_value: str, key_type: str = 'SecureString', overwrite: bool = True):
    """
    Parameter Store에 값을 저장합니다.
    
    Args:
        param_name (str): 저장할 파라미터 이름 (예: '/aws-credential/smtp2go-key')
        param_value (str): 저장할 값
        key_type (str): 파라미터 타입 ('String', 'StringList', 'SecureString')
        overwrite (bool): 이미 존재할 경우 덮어쓸지 여부

    Returns:
        dict: AWS 응답 객체
    """
    ssm = boto3.client('ssm', region_name=region_name)

    try:
        response = ssm.put_parameter(
            Name=param_name,
            Value=param_value,
            Type=key_type,
            Overwrite=overwrite
        )
        return response
    except ClientError as e:
        raise RuntimeError(f"Parameter 저장 실패: {e}")
```

```python
import boto3
import json
from botocore.exceptions import ClientError

region_name = 'ap-northeast-2'  # 원하는 리전으로 수정

def get_parameter_key(param_name='/aws-credential/smtp2go-key'):
    """
    Systems Manager Parameter Store 에서 SecureString 타입의 파라미터를 가져옵니다.
    """
    try:
        ssm = boto3.client('ssm', region_name=region_name)
        response = ssm.get_parameter(
            Name=param_name,
            WithDecryption=True  # SecureString일 경우 True 필수
        )
        return response['Parameter']['Value']
    except ClientError as e:
        raise RuntimeError(f"Parameter Store 접근 중 오류 발생: {e}")
```
